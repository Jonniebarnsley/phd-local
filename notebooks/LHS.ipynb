```python
from local.lib.latinhypercube import LHS_extendable
from skopt.space import Space
import matplotlib.pyplot as plt
```

```python
lhs = LHS_extendable(criterion="maximin")
space = Space([(0., 1.), (0., 1.)])
```

```python
samples = lhs.generate(space.dimensions, 30)
x = list(zip(*samples))[0]
y = list(zip(*samples))[1]
plt.scatter(x, y, marker='o', color='white', label='LHS samples', edgecolors='indianred')
plt.xlabel('Parameter A')
plt.ylabel('Parameter B')

plt.gca().set_aspect('equal')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.title('Perturbed Parameter Ensemble')

plt.gcf().set_dpi(200)
```
