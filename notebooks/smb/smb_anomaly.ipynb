```python
import xarray as xr
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
home = Path("/Volumes/LaCie/smb/smb_anomaly_timeseries_antarctic_mask")
ssp126 = home / "ssp126"
ssp585 = home / "ssp585"
ssp534over = home / "ssp534-over"
```

```python
fig, ax = plt.subplots(figsize=(12, 6), dpi=200)
for file in ssp126.glob("*.nc"):
    model = file.stem.split("_")[2]
    ds = xr.open_dataset(file)
    ax.plot(ds.time, ds.smb_anomaly, label=model, lw=0.5)
plt.legend()
plt.title("SSP126 Antarctic SMB Anomaly w.r.t 1995-2014 climatology")
```
