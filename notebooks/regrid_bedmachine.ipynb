### Correcting bed elevation over lake Vostok

In most parts of the ice sheet, BedMachine's surface elevation is equal to the sum of its bed elevation and ice thickness. The notable exceptions are the ice shelves, where the surface elevation is determined by hydrostatic equilibrium, and subglacial lakes, where the surface elevation is equal to the sum of the bed elevation, ice thickness, and lake depth. Lake Vostok is the largest subglacial lake in Antarctica, and the BISICLES surface elevation there will be too low if we naïvely input the Bedmachine bed elevation and thickness (BISICLES doesn't know about the lake water, so it will dump ice into the lake instead). We will apply a correction to the bed elevation fix this.

```python
vostok = bedmachine.mask == 4 # Lake Vostok mask
corrected_bed = np.where(vostok, bedmachine.surface - bedmachine.thickness, bedmachine.bed)
bedmachine['corrected_bed'] = (('y', 'x'), corrected_bed)
bedmachine.corrected_bed.attrs = bedmachine.bed.attrs
bedmachine.corrected_bed.attrs['long_name'] = 'Corrected bed elevation'
bedmachine.corrected_bed.attrs['description'] = 'Bed elevation corrected over Lake Vostok'
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(12, 5))

ice_density = 917.  # kg/m^3
ocean_density = 1027.  # kg/m^3
grounded = bedmachine.thickness > -bedmachine.bed * ocean_density / ice_density

uncorrected_surface_elevation = (bedmachine.bed + bedmachine.thickness).where(grounded)
corrected_surface_elevation = (bedmachine.corrected_bed + bedmachine.thickness).where(grounded)

uncorrected_surface_elevation.plot(ax=ax[0])
corrected_surface_elevation.plot(ax=ax[1])

for a in ax:
    a.set_aspect('equal')

ax[0].set_title('Uncorrected surface elevation')
ax[1].set_title('Corrected surface elevation over Lake Vostok')
plt.tight_layout()
```

```python
measures
# MEaSUREs isn't big enough to fit BISICLES extent. Need to pad.
```

```python
# pad measures until it exceeds the extent of bisicles
N = 1000 # any big number will do
padded = measures.pad(x=(N, N), y=(N, N))

x = measures.x
dx = x[1] - x[0]
padded_axis_vals = np.arange(x.min()-N*dx, x.max()+N*dx+1, dx)
padded['x'] = padded_axis_vals
padded['y'] = -1*padded_axis_vals
padded
```

```python
measures_1km = padded.interp(x=centered.x, y=centered.y, method='nearest')
measures_1km
```

```python
speed = np.sqrt(measures_1km.VX**2 + measures_1km.VY**2)
speed.plot(norm = mcolors.LogNorm(vmin=0.1))
```

```python
velc = (measures_1km.VX.fillna(0) != 0).astype(int)
velc.plot()
```

```python
ds = xr.Dataset(
    data_vars = {
        'thk': bedmachine_1km.thickness,
        'topg': bedmachine_1km.corrected_bed,
        'c_third': centered.c_third,
        'xvel': measures_1km.VX,
        'yvel': measures_1km.VY,
        'velc': velc
    },

    coords={
        'x': centered.x,
        'y': centered.y
    }
)

ds = ds.fillna(0)

# return to original bisicles grid (not centered on south pole)
ds['x'] = bisicles.x
ds['y'] = bisicles.y

ds.to_netcdf('~/Downloads/inversion_inputs.nc')
```
