```python
import matplotlib.pyplot as plt
from pathlib import Path
from local.lib import GoelzerSLC as gslc

thkpath = Path('/Users/jonniebarnsley/data/pliocene/thickness')
zbpath = Path('/Users/jonniebarnsley/data/pliocene/Z_base')

slc = gslc.getEnsembleSLC(thkpath, zbpath, parallel=True)
```

```python
slc
```

```python
for run in slc.run:
    plt.plot(slc.time, slc.sel(run=run), color='black', lw=0.5, alpha=0.3)

plt.ylabel('Sea level contribution (m)')
plt.xlabel('Time (years)')
plt.ylim(-20, 30)

plt.savefig('/Users/jonniebarnsley/Downloads/pliocene_timeseries.png', dpi=300, bbox_inches='tight')
```
