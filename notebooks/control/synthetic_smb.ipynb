```python
import xarray as xr
from pathlib import Path
```

```python
smb_dir = Path("/Volumes/LaCie/inversions/ppe/0lev/synthetic_smb")
datasets = []
for file in smb_dir.glob("*.nc"):
    ds = xr.open_dataset(file)
    datasets.append(ds)
ensemble = xr.concat(datasets, dim="run")
ensemble = ensemble.assign_coords(run=range(len(datasets)))
ensemble
```

```python
mean_smb = ensemble.smb_synthetic.mean(dim="run")
mean_smb.plot()
```
