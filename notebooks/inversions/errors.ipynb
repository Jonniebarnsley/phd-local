```python
import xarray as xr
import numpy as np
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
from local.lib.myplots import spatialPlot
```

```python
rel1 = xr.open_dataset('/Users/jonniebarnsley/Downloads/rel1_error.nc')
u = rel1.xVel
v = rel1.yVel
speed = np.sqrt(u**2 + v**2)

fig = plt.figure()
ax = plt.axes(projection=ccrs.SouthPolarStereo())
im = spatialPlot(ax, speed, cmap='viridis', vmax=10_000)
fig.colorbar(im)
```

```python
def plotSpeed(file):
    u = file.xVel
    v = file.yVel
    speed = np.sqrt(u**2 + v**2)

    fig = plt.figure()
    ax = plt.axes(projection=ccrs.SouthPolarStereo())
    im = spatialPlot(ax, speed, cmap='viridis', vmax = 10_000)
    fig.colorbar(im)
```

```python
rel2 = xr.open_dataset('/Users/jonniebarnsley/Downloads/rel2_error.nc')
plotSpeed(rel2)
```
