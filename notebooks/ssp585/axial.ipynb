```python
import xarray as xr
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
from local.lib.GoelzerSLC import getEnsembleSLC, GoelzerSLC
from dask.distributed import Client, default_client
from distributed import get_client
```

```python
axial_p = Path('/Users/jonniebarnsley/Downloads/axial')
Joughin_p = Path('/Users/jonniebarnsley/Downloads/joughin')
A0_multiplier_p = Path('/Users/jonniebarnsley/Downloads/A0_multiplier')
C_reg_p = Path('/Users/jonniebarnsley/Downloads/Creg')
long_relax_p = Path('/Users/jonniebarnsley/Downloads/long_relax')
```

```python
def get_slc(ensemble):

    thkpath = ensemble / 'thickness'
    zbpath = ensemble / 'Z_base'

    slc = getEnsembleSLC(thkpath, zbpath, parallel=True)
    return slc
```

```python
axial = get_slc(axial_p)
```

```python
Joughin = get_slc(Joughin_p)
```

```python
Joughin_labels = {
    1: 'J300',
    2: 'J50'
}
```

```python
A0_multiplier = get_slc(A0_multiplier_p)
```

```python
A0_multiplier_labels = {
    1: 'n=2, ref=30 kPa',
    2: 'n=2, ref=50 kPa',
    3: 'n=4, ref=30 kPa',
    4: 'n=4, ref=50 kPa'
}
```

```python
Creg_labels = {
    1: '$a_C$=1e+6',
    2: '$a_C$=1e+4',
    3: '$a_C$=1e-2',
    4: '$a_C$=1e+2',
    5: '$a_C$=1e+8',
}
```

```python
long_relax = get_slc(long_relax_p)
```
