```python
from matplotlib import pyplot as plt
from pathlib import Path

from slepy import SLECalculator
```

```python
data = Path('/Users/Jonniebarnsley/Downloads/high-aPhi')

with SLECalculator() as calc:
    slc = calc.process_ensemble(data / 'thickness', data / 'Z_base')
```

```python
for run in slc.run:
    plt.plot(slc.time, slc.sel(run=run), label=run.values)

plt.legend()
```
