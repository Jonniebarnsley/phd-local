```python
import xarray as xr
from matplotlib import pyplot as plt
from pathlib import Path

from slepy import SLECalculator
```

```python
files = Path('/Users/jonniebarnsley/Downloads/long_relax')
```

```python
with SLECalculator() as calc:
    sle = calc.process_ensemble(files / 'thickness', files / 'Z_base')
```

```python
relax = {
    1: '10 yr',
    2: '15 yr',
    3: '20 yr',
    4: '25 yr',
    5: '30 yr',
}

for run in sle.run:
    plt.plot(sle.time, sle.sel(run=run), label=relax[int(run)])
plt.legend()
plt.title('relaxation length')
plt.ylabel('SLE (m)')
plt.xlabel('time (yr)')
plt.xlim(2000, 2100)
plt.ylim(0, 0.15)
```
