- 16 iterations (previous standard)
- 24 iterations (mid-range)
- 32 iterations (more time/resouce consuming, but potentially more accurate)

```python
from slepy import SLECalculator
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
files = Path('/Users/jonniebarnsley/Downloads/high-reg')
with SLECalculator() as calc:
    sle = calc.process_ensemble(files / 'thickness', files / 'Z_base')
```

```python
for run in sle.run:
    plt.plot(sle.time, sle.sel(run=run), label=int(run))
plt.legend()
```
